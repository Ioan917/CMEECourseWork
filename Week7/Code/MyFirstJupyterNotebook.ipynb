```r
# Change kernel to python3

a = "this is python!"; print(a)
```

```r
# Change kernel to R

a <- "this is R!"; cat(a)
```

```r
# Change kernel back to python3

import matplotlib.pyplot as p
import scipy as sc

x = sc.arange(0, 5, 0.1); y = sc.sin(x)
p.plot(x, y); p.show()
```

```r
# Change kernel back to R

require(ggplot2)
library(repr) # to resize plot with jupyter - package is part of the IRkernel
options(repr.plot.width = 8, repr.plot.height = 8)

x <- seq(0, 5, 0.1); y <- sin(x)
qplot(x, y, geom = "line") # large figure
```
